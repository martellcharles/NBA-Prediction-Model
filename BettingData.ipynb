```python
def addCol(df,colName,szn):
    daList = []
    for i in range(len(df.iloc[:,0])):
        daList.append(szn)
    df.insert(loc=1,column=colName,value=daList)
```

```python
def cleanTeamCol(df):
    teams = list(df['Team'])
    daList = []
    for i in range(len(teams)):
        if teams[i] == 'Orlando':
            daList.append('ORL')
        elif teams[i] == 'Indiana':
            daList.append('IND')
        elif teams[i] == 'Chicago':
            daList.append('CHI')
        elif teams[i] == 'Miami':
            daList.append('MIA')
        elif teams[i] == 'LAClippers' or teams[i] == 'LA Clippers':
            daList.append('LAC')
        elif teams[i] == 'LALakers':
            daList.append('LAL')
        elif teams[i] == 'Philadelphia':
            daList.append('PHI')
        elif teams[i] == 'Brooklyn':
            daList.append('BKN')
        elif teams[i] == 'Cleveland':
            daList.append('CLE')
        elif teams[i] == 'Boston':
            daList.append('BOS')
        elif teams[i] == 'Toronto':
            daList.append('TOR')
        elif teams[i] == 'Washington':
            daList.append('WAS')
        elif teams[i] == 'Detroit':
            daList.append('DET')
        elif teams[i] == 'Milwaukee':
            daList.append('MIL')
        elif teams[i] == 'NewYork':
            daList.append('NYK')
        elif teams[i] == 'Charlotte':
            daList.append('CHA')
        elif teams[i] == 'Houston':
            daList.append('HOU')
        elif teams[i] == 'Minnesota':
            daList.append('MIN')
        elif teams[i] == 'NewOrleans':
            daList.append('NOP')
        elif teams[i] == 'Atlanta':
            daList.append('ATL')
        elif teams[i] == 'Dallas':
            daList.append('DAL')
        elif teams[i] == 'Memphis':
            daList.append('MEM')
        elif teams[i] == 'SanAntonio':
            daList.append('SAS')
        elif teams[i] == 'OklahomaCity' or teams[i] == 'Oklahoma City':
            daList.append('OKC')
        elif teams[i] == 'Utah':
            daList.append('UTA')
        elif teams[i] == 'Portland':
            daList.append('POR')
        elif teams[i] == 'Phoenix':
            daList.append('PHX')
        elif teams[i] == 'Denver':
            daList.append('DEN')
        elif teams[i] == 'Sacramento':
            daList.append('SAC')
        elif teams[i] == 'GoldenState' or teams[i] == 'Golden State':
            daList.append('GSW')
    df['Team'] = daList
    return df
```

```python
def addGameIDandDate(df):
    gameIdList = []
    dateList = []
    for index, row in df.iterrows():
        date = str(row[0])
        if len(date) == 4:
            month = date[:2]
            day = date[2:]
        elif len(date) == 3:
            month = '0'+date[:1]
            day = date[1:]
        season = row[1]
        team = row[4]
        if int(month) > 9:
            year = season[:2]
        else:
            year = season[3:]
        date = '20' + year + '-' + month + '-' + day
        gameId = date + team
        dateList.append(date)
        gameIdList.append(gameId)
    df.insert(loc=2,column='GameID',value=gameIdList)
    df['Date'] = dateList
    return df
```
